# CSV to JSON Converter

This notebook converts CSV files to JSON format.

```python
import pandas as pd
import json
import os
```

```python
# Specify the CSV file path
csv_file = "airline_db_merged_translated.xlsx - users.csv"

# Read the CSV file
df = pd.read_csv(csv_file)

print(f"Loaded {len(df)} rows from {csv_file}")
df.head()
```

```python
# Convert DataFrame to JSON format like db.json
# Assumes first column contains the unique ID/key
id_column = df.columns[0]
print(f"Using '{id_column}' as the key column")

# Convert to dictionary with IDs as keys
data_dict = {}
for _, row in df.iterrows():
    key = str(row[id_column])
    # Convert row to dict, excluding the ID column from values
    row_dict = row.drop(id_column).to_dict()
    data_dict[key] = row_dict

# Determine the top-level key based on filename
if "users" in csv_file.lower():
    json_structure = {"users": data_dict}
elif "flights" in csv_file.lower():
    json_structure = {"flights": data_dict}
elif "reservations" in csv_file.lower():
    json_structure = {"reservations": data_dict}
else:
    json_structure = {"data": data_dict}

print(f"\nJSON structure preview:")
print(f"Top-level key: {list(json_structure.keys())[0]}")
print(f"Number of entries: {len(data_dict)}")
print(f"Sample keys: {list(data_dict.keys())[:3]}")
```

```python
# Save to JSON file
output_file = csv_file.replace(".csv", ".json")

with open(output_file, "w", encoding="utf-8") as f:
    json.dump(json_structure, f, indent=2, ensure_ascii=False)

print(f"Successfully saved JSON to: {output_file}")
print(f"Structure: {list(json_structure.keys())[0]} with {len(data_dict)} entries")
```
