```python
from faker import Faker

def generate_person_names(locale_code, count=10):
    """
    Generates a list of unique fake male and female names for a given locale.

    Args:
        locale_code (str): The locale code for name generation (e.g., 'en_US', 'th_TH').
        count (int): The number of names to generate for each gender.

    Returns:
        tuple: A tuple containing two lists: (female_names, male_names).
    """
    # Create a new Faker instance for the specific locale to ensure proper unique tracking
    # and locale-specific generation.
    local_fake = Faker(locale_code)

    female_names = []
    male_names = []

    # Generate unique female names
    for _ in range(count):
        female_names.append(local_fake.unique.first_name_female() + ' ' + local_fake.unique.last_name_female())

    # Generate unique male names
    for _ in range(count):
        male_names.append(local_fake.unique.first_name_male() + ' ' + local_fake.unique.last_name_male())

    return female_names, male_names
```

```python
# Generate place
@dataclass
class PlaceAddress:
    """
    A raw address returned by the Places API.
    """
    name: str
    formatted_address: str
    lat: Optional[float] = None
    lng: Optional[float] = None

def fetch_addresses(country: str, count: int = 5) -> list[PlaceAddress]:
    """
    Fetch real address from a given country using Google Places API (text search)

    Args:
        country : Must be a key in COUNTRIES
        count   : How many addresses to fetch. Max 20 per API call.

    Returns:
        List of PlaceAddress objects.

    Raises:
        ValueError      : If the country is not in COUNTRIES.
        requests.HTTPError : If the API returns a non-200 status.
    """
    if country not in COUNTRIES:
        raise ValueError(
            f"'{country}' is not configured. "
            f"Available: {list(COUNTRIES.keys())}"
        )

    region_code, query = COUNTRIES[country]

    response = requests.post(
        PLACES_ENDPOINT,
        json={
            "textQuery": query,
            "regionCode": region_code,
            "languageCode": "en",
            "pageSize": min(count, 20),
        },
        headers={
            "Content-Type": "application/json",
            "X-Goog-Api-Key": API_KEY,
            "X-Goog-FieldMask": (
                "places.displayName,"
                "places.formattedAddress,"
                "places.location" # for validation
            ),
        },
        timeout=REQUEST_TIMEOUT,
    )

    if not response.ok:
        print(f"  [Places API] Error {response.status_code}: {response.text}")
        response.raise_for_status()

    places = response.json().get("places", [])

    return [
        PlaceAddress(
            name=p.get("displayName", {}).get("text", "Unknown"),
            formatted_address=p.get("formattedAddress", ""),
            lat=p.get("location", {}).get("latitude"),
            lng=p.get("location", {}).get("longitude"),
        )
        for p in places
        if p.get("formattedAddress")    # skip any place with no address
    ]
```

```python
for country in COUNTRIES:
    print(f"Fetching addresses for {country}...")
    fetched_batch = fetch_addresses(country, count=5)
    print(f"  Fetched {len(fetched_batch)} addresses.")
```

```python
target_valid_count = 10
max_fetch_attempts = 3 # Limit the number of API calls per country to prevent excessive usage

for country in COUNTRIES:
    print(f"Fetching addresses for {country}...")

    # Use a set to store unique formatted addresses to avoid duplicates efficiently
    # and a list to store the PlaceAddress objects in order of finding.
    unique_formatted_addresses = set()
    found_valid_addresses = []

    attempts = 0
    while len(found_valid_addresses) < target_valid_count and attempts < max_fetch_attempts:
        # We need to fetch enough to potentially get the remaining valid addresses.
        num_to_request = 20

        fetched_batch = fetch_addresses(country, count=num_to_request)

        newly_valid_from_batch = [addr for addr in fetched_batch if addr.lat and addr.lng]

        for addr in newly_valid_from_batch:
            if addr.formatted_address not in unique_formatted_addresses:
                found_valid_addresses.append(addr)
                unique_formatted_addresses.add(addr.formatted_address)

            if len(found_valid_addresses) >= target_valid_count:
                break # Stop if we have enough valid addresses

        if len(found_valid_addresses) < target_valid_count:
            print(f"  Only {len(found_valid_addresses)} valid addresses found. Retrying (attempt {attempts + 1}/{max_fetch_attempts})...")

        attempts += 1

    print(f"  Finished processing {country}. Found {len(found_valid_addresses)} valid addresses (total API calls: {attempts}).")
```

## Other method: Faker + Geocoding

```python
API_KEY = "AIzaSyDhzvWrgYS-7i_-XvajAO3bspdoZCEbzZU"
VALIDATION_ENDPOINT = "https://maps.googleapis.com/maps/api/geocode/json"
```

```python
import requests

def validate_address(address: str, region_code: str) -> dict:
    response = requests.get(
        VALIDATION_ENDPOINT,
        params={
            "address": address,
            "region": region_code.lower(),
            "key": API_KEY,
        }
    )

    data = response.json()

    # https://developers.google.com/maps/documentation/geocoding/guides-v3/requests-geocoding#StatusCodes
    if data["status"] != "OK":
        return {"verdict": "FIX", "reason": data["status"]}

    top = data["results"][0]
    location_type = top["geometry"]["location_type"]
    partial_match = top.get("partial_match", False)

    if location_type in ("ROOFTOP", "RANGE_INTERPOLATED"):
        verdict = "ACCEPT"
    elif location_type == "GEOMETRIC_CENTER":
        verdict = "CONFIRM"
    else:
        verdict = "FIX"

    return {
        "verdict": verdict,
        "location_type": location_type,
        "partial_match": partial_match,
        "formatted_address": top["formatted_address"],
        "lat": top["geometry"]["location"]["lat"],
        "lng": top["geometry"]["location"]["lng"],
    }


# Test it
result = validate_address("1600 Amphitheatre Pkwy, Mountain View, CA", "US")
print(result['verdict'])
```

https://developers.google.com/maps/documentation/geocoding/guides-v3/requests-geocoding

* partial_match indicates that the geocoder did not return an exact match for the original request, though it was able to match part of the requested address. You may wish to examine the original request for misspellings and/or an incomplete address.

* location_type stores additional data about the specified location. The following values are currently supported:

1.   "ROOFTOP" indicates that the returned result is a precise geocode for which we have location information accurate down to street address precision.

2.   "RANGE_INTERPOLATED" indicates that the returned result reflects an approximation (usually on a road) interpolated between two precise points (such as intersections). Interpolated results are generally returned when rooftop geocodes are unavailable for a street address.

3. "GEOMETRIC_CENTER" indicates that the returned result is the geometric center of a result such as a polyline (for example, a street) or polygon (region). -> e.g. street exists but house number is missing/unrecognized; point to whole street rather than specific point; partically correct

4. "APPROXIMATE" indicates that the returned result is approximate.

```python
from faker import Faker

def generate_and_validate(country: str, count: int = 10, max_retries: int = 5):
    locale, region_code = COUNTRY_LOCALES[country]
    fake = Faker(locale)

    print(f"\n{country}:")
    for i in range(count):
        verdict = "FIX" # Initialize verdict to ensure the loop runs at least once
        address = ""
        retries = 0

        while verdict in ("FIX", "CONFIRM") and retries < max_retries:
            address = fake.address().replace("\n", ", ")  # flatten multiline
            result = validate_address(address, region_code)
            verdict = result['verdict']
            if verdict not in ("FIX", "CONFIRM"):
                # If it's ACCEPT or any other final verdict, break out
                break
            retries += 1
            # Optional: print retry attempts
            # print(f"    [Retry {retries}/{max_retries}] Verdict: {verdict}, Address: {address}")

        print(f"  [{result['verdict']}] {address}")
```
